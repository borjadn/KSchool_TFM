```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('vgsales_info.csv', index_col = 0)
data.head()
```

Let's add format to the sales figures and the dates.

We can see that the `XXXX_Sales` column uses `m` as the million unit. So we select the Sales columns, strip the `m` and proceed to convert it to a numerical value

```python
sales_col = ['Total_Shipped','Total_Sales','NA_Sales','EU_Sales','JP_Sales','Other_Sales']

for col in sales_col:
    data[col] = data[col].str.strip('m')
    data[col] = data[col].astype('float') * 1000000
```

```python
data.sample(5)
```

---
Now we proceed to clean the `Release` and `Last_Update` columns.

We can observe the following:

- There are NaNs
- Days are in the ~st ~nd ~rd ~th format
- Months are in 3 letter format
- Years have the last 2 digits format

```python
years = []

for col in ['Release','Last_Update']:
       
        for row in data[col].str.split(' '):
            
            if type(row) != float and (not row[2] in years):
                
                years.append(row[2])
                

pd.Series(years).sort_values(ignore_index = True, ascending = False)
```

We can see that:
- The years belonging to the 21th century range between 2000 and 2021
- The years belonging to the 20th century range between 1970 and 1999

Now we are going to split the string into the different elements to transform them into a more intelligible format and merge them later. 

```python
months = {'Jan':'01', 'Feb':'02', 'Mar':'03', 'Apr':'04', 'May':'05', 'Jun':'06',
          'Jul':'07', 'Aug':'08', 'Sep':'09', 'Oct':'10', 'Nov':'11', 'Dec':'12'}
```

```python
for col in ['Release','Last_Update']:

        data[col] = data[col].str.split(' ')
        
        for row in data[col]:
            
            if type(row) != float:
                
                row[0] = row[0].strip('stndrh') # Days
                
                row[1] = months[row[1]]         # Months
                
                if int(row[2]) < 22:            # Years
                    
                    row[2] = '20' + row[2]      # 21th century
                    
                else:
                    
                    row[2] = '19' + row[2]      # 20th century
        
        
        data[col] = data[col].str.join('-')   # Merge the str array into a single string in the dd-mm-YYYY format
```

```python
data.sample(10)
```

---
Now that there's nothing left to clean (in the formatting sense), let's check for NaN values to see what to do with them (as we have already seen many).

```python
data.isna().sum()
```

There are so many NaNs in the dataframe we have to decide what to do with them!

We are interested in the sales, so taking non-NaN `Total_Sales` values seems logical (other Sales columns have more NaNs, so it won't affect as much).

As we are going to merge this dataset with other datasets, we could even drop some columns we won't use!:

- We won't need these many NaN score columns (`VGC_Score`, `Critic_Score`, `User_Score`) as we will try to scrape data from Metacritic.
- We won't need the `Total_Shipped` column either as we are interested in sales, not in shipping

```python
data_filtered = data[data['Total_Sales'].notna()].drop(['VGC_Score','Critic_Score','User_Score'], axis = 1)
data_filtered.sample(10)
```

```python
data_filtered.to_csv("vgsales_clean.csv", sep=",", encoding='utf-8')
```
