# Web Scraping RAWG
We will define a function to scrape this web site, as we had some errors interrupting the process.

We are going to scrape the data until no pages are left.

```python
api_key =  # Introduce your API key here

url = 'https://api.rawg.io/api/games?key=' + api_key + '&page='
```

```python
hasNextPage = True # Condition to stop the while loop
page_n = 1 # Global page number count
games = pd.DataFrame() # Final dataframe
```

```python
def scrape(url, df):
    
    '''
    Given an url and a dataframe, it will scrape the data from the web site we gave the url and concatenate the result to the dataframe.
    url : URL to the RAWG API. We have to specify the key we get from RAWG.
    '''
    
    while hasNextPage:
        
        response = requests.get(url+str(page_n)).json()

        aux = pd.DataFrame.from_dict(response['results'], orient = 'columns')

        df = pd.concat([df, aux])

        print('Page: ' + str(page_n))

        page_n += 1

        if response['next'] != (url + str(page_n)):

            hasNextPage = False

    print('Finished') 
```
